```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from wordcloud import WordCloud, STOPWORDS
import re

def words(df):
    corpus = '' 
    stopwords = set(STOPWORDS) 
    stopwords.add('amp')
    stopwords.add('s');

    # iterate through the tweets
    for text in df.Tweet: 

        # remove url links in tweets
        text = result = re.sub(r"http\S+", "", str(text))
        # split the value 
        tokens = [word.lower() for word in word_tokenize(text)]
        loweralphabets="abcdefghijklmnopqrstuvwxyz"
        upperalphabets="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        x=loweralphabets+upperalphabets
        res=[]
        for i in tokens:
            a=""
            for j in i:
                if j in x:
                    a+=j
            res.append(a)
        # adding the tokenized words into corpus
        corpus += " ".join(res)+" "

    wordcloud = WordCloud(width = 800, height = 800, 
                    background_color = "white", 
                    stopwords = stopwords, 
                    min_font_size = 10).generate(corpus) 

    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 

    plt.show() 
```

```python
words(modidf)
```

```python
words(rahuldf)
```

```python
import re
def cleantext(text):
    text = re.sub(r"@[A-Za-z0-9]+", "", text) # Remove Mentions
    text = re.sub(r"#", "", text) # Remove Hashtags Symbol
    text = re.sub(r"RT[\s]+", "", text) # Remove Retweets
    text = re.sub(r"https?:\/\/\S+", "", text) # Remove The Hyper Link
    
    return text
# Clean The Text
modidf["Tweet"] = modidf["Tweet"].apply(cleantext)
modidf.head()
```

```python
from textblob import TextBlob
# Get The Subjectivity
def sentiment_analysis(ds):
    sentiment = TextBlob(ds["Tweet"]).sentiment
    return pd.Series([sentiment.subjectivity, sentiment.polarity])
# Adding Subjectivity & Polarity
modidf[["subjectivity", "polarity"]] = modidf.apply(sentiment_analysis, axis=1)
modidf
```

```python
def analysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
# Create a New Analysis Column
modidf["analysis"] = modidf["polarity"].apply(analysis)
# Print The Data
modidf
```

```python
import matplotlib.pyplot as plt
s = modidf["analysis"].value_counts()
plt.pie(s, labels=s.index, autopct='%1.1f%%')
plt.show()
```

```python
rahuldf["Tweet"] = rahuldf["Tweet"].apply(cleantext)
rahuldf[["subjectivity", "polarity"]] = rahuldf.apply(sentiment_analysis, axis=1)
rahuldf["analysis"] = rahuldf["polarity"].apply(analysis)
s = rahuldf["analysis"].value_counts()
plt.pie(s, labels=s.index, autopct='%1.1f%%')
plt.show()
```
